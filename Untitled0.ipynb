```python


file_path = '/content/drive/MyDrive/Big Data Lanjut s5/Uas/archive/heart.csv'
data = pd.read_csv(file_path)

# Menampilkan data
data.head()
```

```python
# Preprocessing
# 1. Rename columns for consistency
data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]

# 2. Check and handle missing values
data.fillna(data.mean(), inplace=True)

# 3. Convert categorical variables to numeric using one-hot encoding
# Assuming categorical columns: 'sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal'
data = pd.get_dummies(data, columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal'], drop_first=True)

# 4. Summary of processed data
print("Processed DataFrame Info:")
print(data.info())
print("\nFirst 5 rows of the processed data:")
print(data.head())
```

```python
# Spark Session
spark = SparkSession.builder.appName("HeartDiseaseClassification").getOrCreate()
```

```python
# Convert to Spark DataFrame
spark_df = spark.createDataFrame(data)
```

```python
# Assembling features
feature_cols = [col for col in spark_df.columns if col != 'target']
assembler = VectorAssembler(inputCols=feature_cols, outputCol='features')
spark_df = assembler.transform(spark_df)
```

```python
# Split data
train_data, test_data = spark_df.randomSplit([0.8, 0.2], seed=42)
```

```python
# Initialize models
rf_model = RandomForestClassifier(labelCol='target', featuresCol='features', seed=42)
gbt_model = GBTClassifier(labelCol='target', featuresCol='features', seed=42)
lr_model = LogisticRegression(labelCol='target', featuresCol='features', maxIter=10)
nb_model = NaiveBayes(labelCol='target', featuresCol='features')
```

```python
# Train models
rf_fit = rf_model.fit(train_data)
gbt_fit = gbt_model.fit(train_data)
lr_fit = lr_model.fit(train_data)
nb_fit = nb_model.fit(train_data)
```

```python
# Evaluate models
evaluator_roc = BinaryClassificationEvaluator(labelCol='target', metricName='areaUnderROC')
evaluator_acc = MulticlassClassificationEvaluator(labelCol='target', metricName='accuracy')
evaluator_f1 = MulticlassClassificationEvaluator(labelCol='target', metricName='f1')
evaluator_precision = MulticlassClassificationEvaluator(labelCol='target', metricName='weightedPrecision')
evaluator_recall = MulticlassClassificationEvaluator(labelCol='target', metricName='weightedRecall')
```

```python
# Evaluate and store metrics
metrics = {}
for name, model in zip(['Random Forest', 'GBT', 'Logistic Regression', 'Naive Bayes'], [rf_fit, gbt_fit, lr_fit, nb_fit]):
    predictions = model.transform(test_data)
    metrics[name] = {
        'AUC': evaluator_roc.evaluate(predictions),
        'Accuracy': evaluator_acc.evaluate(predictions),
        'F1 Score': evaluator_f1.evaluate(predictions),
        'Precision': evaluator_precision.evaluate(predictions),
        'Recall': evaluator_recall.evaluate(predictions)
    }
```

```python
# Print metrics
for model_name, model_metrics in metrics.items():
    print(f"\nModel: {model_name}")
    for metric_name, metric_value in model_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")
```
